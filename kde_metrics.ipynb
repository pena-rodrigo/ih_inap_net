```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#matplotlib inline
```

```python
case1 = pd.read_csv('case1.dat',names=['Time [ms]','Amplitude'],sep=' ')
case2_pk1 = pd.read_csv('case2pk1.dat',names=['Time [ms]','Amplitude'],sep=' ')
case2_pk2 = pd.read_csv('case2pk2.dat',names=['Time [ms]','Amplitude'],sep=' ')
```

```python
case1['time']
```

```python
sns.set(style="white", color_codes=True)
# matplotlib.rc("legend", fontsize=18)
sns.set(font_scale = 1.5)
with sns.axes_style("white"):
    g=sns.jointplot(x=case1['Time [ms]'],y=case1['Amplitude'],kind='kde',xlim=(0,12),ylim=(0,350),stat_func=None) #,xlim=(0,12)
# plt.show()
plt.tick_params(labelsize=18)
# plt.legend(fontsize=18)
g.savefig('./pk1.pdf', format='pdf', dpi=600)
```

```python
sns.set(style="white", color_codes=True)
# matplotlib.rc("legend", fontsize=18)
sns.set(font_scale = 1.5)
with sns.axes_style("white"):
    g=sns.jointplot(x=case2_pk1['time'],y=case2_pk1['amp'],kind='kde',xlim=(0,10),ylim=(0,350),stat_func=None) #,ylim=(0,150)
# plt.show()
plt.tick_params(labelsize=18)
plt.legend(fontsize=18)
g.savefig('./pk2_1.pdf', format='pdf', dpi=600)
```

```python
sns.set(style="white", color_codes=True)
# matplotlib.rc("legend", fontsize=18)
sns.set(font_scale = 1.5)
with sns.axes_style("white"):
    g=sns.jointplot(x=case2_pk2['time'],y=case2_pk2['amp'],kind='kde',xlim=(0,10),ylim=(0,350),stat_func=None) #xlim=(0,12),ylim=(0,150),
# plt.show()
plt.tick_params(labelsize=18)
plt.legend(fontsize=18)
g.savefig('./pk2_2.pdf', format='pdf', dpi=600)
```

```python
case2 = pd.concat([case2_pk1,case2_pk2])
```
